# **Graduation Project -- Zara Gross Margin Analysis**

# **Business Problem Overview**

Zara, a leading global fashion retailer, has consistently maintained a competitive edge in the fast-fashion industry by combining trend-driven designs with quick turnaround times. However, like many large retailers, Zara faces ongoing challenges in maintaining profitability, particularly when it comes to its Gross Margin. Gross Margin, which represents the difference between sales revenue and the cost of goods sold (COGS), is a crucial indicator of the company's profitability and operational efficiency.

Despite Zara’s success, rising production costs, global economic pressures, and an increasingly competitive landscape are putting pressure on its ability to maintain healthy margins. Additionally, the fast-paced nature of the fashion industry, with its reliance on frequent inventory turnover, requires Zara to balance between keeping costs low and meeting customer demand for trendy, high-quality products.

The business problem centers around Zara’s ability to optimize its Gross Margin by effectively managing costs, production efficiency, and the alignment of its product offerings with consumer demand. If Zara fails to sustain or increase its Gross Margin, it could face reduced profitability, decreased investor confidence, and the risk of losing market share to both traditional and emerging competitors in the retail space.

Addressing this problem requires a thorough understanding of the factors that influence Zara’s Gross Margin, including production costs, pricing strategies, inventory management, and market trends. By identifying opportunities for cost reduction, operational improvements, and pricing optimization, Zara can continue to thrive in a highly competitive market while maximizing its profitability.

# **Objective:**

In this project, we aim to analyze Zara's financial and operational data to:

1.Understand Gross Margin Trends: Identify key patterns and fluctuations in Zara's Gross Margin over time and uncover the primary drivers behind these changes.

2.Optimize Cost Structure: Spot inefficiencies in Zara’s supply chain and production processes that could help improve cost management and enhance profitability.

3.Enhance Pricing and Inventory Strategy: Provide actionable insights to optimize Zara's pricing strategies and inventory management, leading to better margins and long-term growth.

# **Business Impact:**

This analysis will empower Zara to:
1.Enhance Profitability: By optimizing gross margin trends and identifying cost-saving opportunities in Zara's supply chain, this project can significantly improve profitability.

2.Boost Operational Efficiency: By uncovering inefficiencies in production, inventory management, and store operations, Zara can streamline its processes, leading to faster turnaround times, better stock management, and cost reductions.

3.Strengthen Competitive Positioning: By optimizing pricing strategies, inventory allocation, and store performance, Zara can improve its market positioning, offer better value to customers, and respond more dynamically to market trends.

4.Optimize Store Management: By analyzing sales data and store performance, Zara can tailor its inventory distribution, adjust store layouts, and implement effective staffing strategies to maximize in-store sales and customer satisfaction.

By addressing cost management, margin optimization, and store operations, this project contributes to Zara’s broader goal of sustaining long-term growth, enhancing profitability, and maintaining its competitive edge in the fast-fashion industry. 
# **2.Changing Data Types for Efficient Data Analysis:**

```python
# --- Calculating Gross Margin per Transaction Line ---

import pandas as pd

print("--- Starting Gross Margin Calculation ---")

if 'df_merged' in locals() and isinstance(df_merged, pd.DataFrame):
    print("Calculating 'Gross Margin' column in df_merged...")


    required_cols = ['Unit Price', 'Quantity', 'Discount', 'Production Cost']
    if all(col in df_merged.columns for col in required_cols):
        try:

            df_merged['Gross Revenue'] = df_merged['Unit Price'] * df_merged['Quantity']
            df_merged['Net Revenue'] = df_merged['Gross Revenue'] * (1 - df_merged['Discount'])

            # Calculate Cost of Goods Sold (COGS)
            df_merged['COGS'] = df_merged['Production Cost'] * df_merged['Quantity']

            # Calculate Gross Margin
            df_merged['Gross Margin'] = df_merged['Net Revenue'] - df_merged['COGS']
            print("'Gross Margin', 'Gross Revenue', 'Net Revenue', and 'COGS' columns calculated successfully.")

            print("\n--- Head of df_merged with new columns ---")
            # Display the head with the new columns
            print(df_merged[['Unit Price', 'Quantity', 'Discount', 'Production Cost', 'Gross Revenue', 'Net Revenue', 'COGS', 'Gross Margin']].head())

            print("\n--- Summary Statistics for Gross Margin ---")
            # Display summary statistics for the new Gross Margin column
            print(df_merged['Gross Margin'].describe())

        except KeyError as e:
            print(f"Error: Required column for calculation not found: {e}")
            print(f"Please ensure the merged DataFrame contains all necessary columns: {required_cols}")
        except Exception as e:
            print(f"An unexpected error occurred during Gross Margin calculation: {e}")

            #  data type errors if columns are not numeric
            if "could not convert string to float" in str(e).lower() or "unsupported operand type" in str(e).lower():
                 print("Tip: Ensure the columns used in the calculation ('Unit Price', 'Quantity', 'Discount', 'Production Cost') are numeric data types (int, float). Check df_merged.info().")

    else:
        missing = [col for col in required_cols if col not in df_merged.columns]
        print(f"Error: Merged DataFrame is missing required columns for calculation: {missing}")
        print("Please review the merge step to ensure these columns were included.")

else:
    print("df_merged DataFrame is not available. Please ensure the merge step was completed successfully.")

print("\n" + "="*70 + "\n")
```

```python
# --- Analysis 6  Data Transformation (Time Features) and Aggregation ---

import pandas as pd
print("--- Time Features & Aggregating Data by Time ---")

required_cols_time = ['Gross Margin', 'Date']
if 'df_merged' in locals() and isinstance(df_merged, pd.DataFrame) and all(col in df_merged.columns for col in required_cols_time):
    print("Preparing time features and aggregating Gross Margin by time...")

    if not pd.api.types.is_datetime64_any_dtype(df_merged['Date']):
        print("Converting 'Date' column to datetime format...")

        df_merged['Date'] = pd.to_datetime(df_merged['Date'], errors='coerce')
        print("Date column converted to datetime format.")
        if df_merged['Date'].isnull().any():
            print("Warning: Some date values could not be converted and were set to NaT.")
            print("Rows with invalid dates might be removed for time analysis.")

            initial_rows = len(df_merged)
            df_merged.dropna(subset=['Date'], inplace=True)
            if len(df_merged) < initial_rows:
                print(f"Removed {initial_rows - len(df_merged)} rows with invalid date values.")
    else:
        print("'Date' column is already in datetime format.")


    if pd.api.types.is_datetime64_any_dtype(df_merged['Date']) and not df_merged['Date'].empty:
        # Extract time features: Year, Month, Day of Week, Hour
        df_merged['Year'] = df_merged['Date'].dt.year
        df_merged['Month'] = df_merged['Date'].dt.month
        # Get month name for better visualization labels
        df_merged['Month Name'] = df_merged['Date'].dt.strftime('%B')
        df_merged['Day of Week'] = df_merged['Date'].dt.dayofweek
        # Get day of week name for better visualization labels
        df_merged['Day Name'] = df_merged['Date'].dt.strftime('%A')
        df_merged['Hour'] = df_merged['Date'].dt.hour
        print("Extracted time features: Year, Month, Month Name, Day of Week, Day Name, Hour.")

        # --- Aggregate Gross Margin by different time diff zone ---

        # Daily Gross Margin (sum and mean)
        print("\nAggregating Daily Gross Margin...")
        daily_gm = df_merged.groupby(df_merged['Date'].dt.date)['Gross Margin'].agg(['sum', 'mean'])
        daily_gm.index = pd.to_datetime(daily_gm.index)
        print("Head of Daily Gross Margin:")
        print(daily_gm.head())

        # Monthly Gross Margin (sum and mean)
        print("\nAggregating Monthly Gross Margin (by Year and Month)...")
        # group by Year and Month number to ensure correct order
        monthly_gm = df_merged.groupby([df_merged['Year'], df_merged['Month']])['Gross Margin'].agg(['sum', 'mean'])
        #  combined Year-Month string or datetime index
        monthly_gm.index = pd.to_datetime(monthly_gm.index.map(lambda x: f'{x[0]}-{x[1]}'))
        print("Head of Monthly Gross Margin:")
        print(monthly_gm.head())

        # Gross Margin by Month of Year (sum and mean, for seasonality pattern)
        print("\nAggregating Gross Margin by Month of Year...")
        # group by Month number and Month Name to ensure correct order
        gm_by_month_of_year = df_merged.groupby(['Month', 'Month Name'])['Gross Margin'].agg(['sum', 'mean']).sort_index(level=0)
        #  the Month Name, maintaining order for potential external use
        gm_by_month_of_year.index = gm_by_month_of_year.index.get_level_values('Month Name')
        print("Gross Margin by Month of Year:")
        print(gm_by_month_of_year)


        # Gross Margin by Day of Week
        print("\nAggregating Gross Margin by Day of Week...")
        # Group by Day of Week number and Day Name to ensure correct order
        gm_by_day_of_week = df_merged.groupby(['Day of Week', 'Day Name'])['Gross Margin'].agg(['sum', 'mean']).sort_index(level=0)
        # Set the index to only the Day Name, maintaining order for potential external use
        gm_by_day_of_week.index = gm_by_day_of_week.index.get_level_values('Day Name')
        print("Gross Margin by Day of Week:")
        print(gm_by_day_of_week)


        # Gross Margin by Hour (sum and mean, for daily pattern)
        print("\nAggregating Gross Margin by Hour...")
        # Group by Hour and sort by index (hour number)
        gm_by_hour = df_merged.groupby('Hour')['Gross Margin'].agg(['sum', 'mean']).sort_index()
        print("Gross Margin by Hour:")
        print(gm_by_hour)



    else:

        print("Error: 'Date' column is not valid or contains only missing values after attempted conversion/cleaning. Cannot perform time analysis.")


else:
    # Provide specific error messages if df_merged or initial required columns are missing
    missing = [col for col in required_cols_time if col not in df_merged.columns] if 'df_merged' in locals() and isinstance(df_merged, pd.DataFrame) else required_cols_time
    print(f"Error: Could not perform time analysis. Missing DataFrames or columns: {missing}")
    print("Please ensure df_merged exists and contains 'Gross Margin' and 'Date'.")

print("\n" + "="*70 + "\n") # Separator
```
